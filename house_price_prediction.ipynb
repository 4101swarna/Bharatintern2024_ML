```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/sample_data/housing.csv')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(data.ocean_proximity).astype(int)
```

```python
data = data.join(pd.get_dummies(data.ocean_proximity).astype(int))
data
```

```python
data.drop(columns = ['ocean_proximity'], inplace=True)
```

```python
data.hist(figsize = (16,10))
```

```python
data.corr() #give correlations with other values
```

```python
from matplotlib import colormaps
list(colormaps)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(),annot=True,cmap ="ocean_r")
```

```python
data['total_rooms'] = np.log(data['total_rooms']  + 1) #+1 to avoid 0 value
data['total_bedrooms'] = np.log(data['total_bedrooms']  + 1)
data['population'] = np.log(data['population']  + 1)
data['households'] = np.log(data['households']  + 1)
```

```python
data.hist(figsize = (16,10))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', data=data, hue = 'median_house_value', palette = 'rocket')
```

```python
data['bedroom_ratio'] = data['total_bedrooms']/data['total_rooms']
data['household_rooms'] = data['total_rooms']/data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(),annot=True,cmap ="winter")
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X= data.drop(['median_house_value'], axis =1)
y = data['median_house_value']

X_train,X_test, y_train, y_test =  train_test_split(X,y, test_size=0.2)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
forest.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python

model.score(X_test,y_test)
```
